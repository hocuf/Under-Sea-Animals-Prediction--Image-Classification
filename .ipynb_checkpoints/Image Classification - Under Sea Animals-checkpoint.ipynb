```python
import tensorflow as tf
import os
import cv2
import imghdr

import numpy as np
from matplotlib import pyplot as plt
```

## Remove dodgy images

```python
data_dir = 'data' 
```

```python
image_exts = ['jpeg','jpg', 'bmp', 'png']
```

1-Create the path to the image file.


2-Using the imghdr library, verify that the file has the correct image file extension. If the file extension is not in the image_exts list, delete the image file and print a warning message on the screen.


3-If the image file cannot be read, delete the file and print an error message on the screen.

```python
for image_class in os.listdir(data_dir): 
    for image in os.listdir(os.path.join(data_dir, image_class)):
        image_path = os.path.join(data_dir, image_class, image)
        try: 
            img = cv2.imread(image_path)
            tip = imghdr.what(image_path)
            if tip not in image_exts: 
                print('Image not in ext list {}'.format(image_path))
                os.remove(image_path)
        except Exception as e: 
            print('Issue with image {}'.format(image_path))
            # os.remove(image_path)
```

## Load Data

**This is used to load images from the dataset and create an image dataset using TensorFlow's Keras API. The 'data' argument specifies the directory in which to load the images. For automatic labelling of images, subdirectories in the data directory are used as classes. This method can be used to prepare the data as a TensorFlow dataset to be used later for model training.**

```python
data = tf.keras.utils.image_dataset_from_directory('data')
data
```

```python
#creates a NumPy array iterator object from a TensorFlow dataset object named data.
data_iterator = data.as_numpy_iterator()
data_iterator
```

```python
#code retrieves the next item (here a dataset or batch) from the iterator object.
batch = data_iterator.next() 
batch
```

## Visualization

```python
fig, ax = plt.subplots(ncols=4, figsize=(20,20))
for idx, img in enumerate(batch[0][:4]):
    ax[idx].imshow(img.astype(int))
    ax[idx].title.set_text(batch[1][idx])
```

## Scale Data

```python
data = data.map(lambda x,y: (x/255, y)) # normalizing
```

```python
data.as_numpy_iterator().next()
```

## Split Data

**The first row allocates 70% of the data as the training dataset. The second row creates the validation dataset by skipping to the end of the training dataset and taking 20% of it. The third row creates the test dataset by skipping to the end of the training and validation datasets and taking the remaining 10%.**

```python
train_size = int(len(data)*.7) %70
val_size = int(len(data)*.2)   %20
test_size = int(len(data)*.1)  %10
```

```python
train_size, test_size
```

```python
train = data.take(train_size)
val = data.skip(train_size).take(val_size)
test = data.skip(train_size+val_size).take(test_size)
```

## Deep Learning Modelling & Testing

```python
train
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Dense, Flatten, Dropout
```

```python
model = Sequential()
model.add(Conv2D(16, (3,3), 1, activation='relu', input_shape=(256,256,3)))
model.add(MaxPooling2D())
model.add(Conv2D(32, (3,3), 1, activation='relu'))
model.add(MaxPooling2D())
model.add(Conv2D(16, (3,3), 1, activation='relu'))
model.add(MaxPooling2D())
model.add(Flatten())
model.add(Dense(256, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile('adam', loss=tf.losses.BinaryCrossentropy(), metrics=['accuracy'])
```

```python
model.summary()
```

```python
#It is used to monitor and analyse the performance of the model during training.
logdir='logs'
tensorboard_callback = tf.keras.callbacks.TensorBoard(log_dir=logdir)
```

```python
hist = model.fit(train, epochs=20, validation_data=val, callbacks=[tensorboard_callback])
```

```python
fig = plt.figure()
plt.plot(hist.history['loss'], color='teal', label='loss')
plt.plot(hist.history['val_loss'], color='orange', label='val_loss')
fig.suptitle('Loss', fontsize=20)
plt.legend(loc="upper left")
plt.show()
```

```python
fig = plt.figure()
plt.plot(hist.history['accuracy'], color='teal', label='accuracy')
plt.plot(hist.history['val_accuracy'], color='orange', label='val_accuracy')
fig.suptitle('Accuracy', fontsize=20)
plt.legend(loc="upper left")
plt.show()
```
